```python
elections_nw['candidate'] = elections_nw['candidate'].fillna('other')
elections_df = elections_nw
elections_df.head()
```

```python
#determine percentage of vote share
elections_df['pct_vote'] = elections_df['candidatevotes'] / elections_df['totalvotes']
elections_df.head()
```

```python
#electoral votes by year
import requests
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/United_States_Electoral_College"
r = requests.get(url)
r_html = r.text
soup = BeautifulSoup(r_html, 'lxml')
tables = soup.find_all('table')
table = tables[4]

new_table = pd.DataFrame(columns=range(0,36), index=range(0,56))
table_head = pd.DataFrame(columns=range(0,36), index=range(0,3))
row_marker = 0
for row in table.find_all('tr'):
    column_marker = 0
    columns = row.find_all(['th','td'])
    for column in columns:
        if row_marker < 3:
            table_head.iloc[row_marker,column_marker] = column.text.replace('\n','')
            #print(column.text.replace('\n',''))
            column_marker += 1
        else:
            new_table.iloc[row_marker,column_marker] = column.text.replace('\n','')
            #print(column.text.replace('\n','')+' '+str(row_marker)+','+str(column_marker))
            column_marker += 1
    row_marker += 1
        
new_table = new_table[4:].drop(columns = 0)
new_table.head()
```

```python
table_index = table_head.iloc[1].to_list()
table_index.insert(0, 'State')
index_dict = {}
for i in range(0,len(table_index)):
    index_dict[i+1] = table_index[i]
new_table = new_table.rename(columns = index_dict).reset_index(drop=True).drop(51)
new_table.tail()
```

```python
ec_votes = new_table.iloc[:,np.r_[0,-5:0]]
ec_votes.head()
```

```python
#check state and year. if state & year match state and 'year, set two equal
years = elections_df['year'].unique()
for year in years:
    if year > 1970 and year <= 1980:
        ec_votes[year] = ec_votes["'72'76'80"]
    elif year > 1980 and year <= 1990:
        ec_votes[year] = ec_votes["'84'88"]
    elif year > 1990 and year <= 2000:
        ec_votes[year] = ec_votes["'92'96'00"]
    elif year > 2000 and year <= 2010:
        ec_votes[year] = ec_votes["'04'08"]
    else:
        ec_votes[year] = ec_votes["'12'16'20"]
```

```python
ec_votes = ec_votes.drop(columns = table_index[-7:-2]).set_index('State').rename(index={'D.C.':'District of Columbia'})
ec_votes.head()
```

```python
elections_df = elections_df.reset_index(drop=True)
for i in range(len(elections_df)):
    elections_df.loc[i,'state_electoral_votes'] = ec_votes.loc[elections_df.loc[i,'state'], elections_df.loc[i,'year']]
elections_df[elections_df['state'] == 'Arizona']
```

```python
elections_df.info()
```

```python
elections_df['state_electoral_votes'] = elections_df['state_electoral_votes'].astype(int)
elections_df.info()
```

```python
#elections_df['candidate_evs'] = elections_df.apply(lambda x: (x['state_electoral_votes'] * x['pct_vote'] if x['pct_vote'] >= 0.05 else 0), axis=1)
elections_df['candidate_evs'] = elections_df['state_electoral_votes'] * elections_df['pct_vote'] 
elections_df.head(20)
```

```python
elections_df['candidate_evs_round'] = round(elections_df['candidate_evs'],0)
elections_df.head(20)
```

```python
elections_non_mp = elections_df[(elections_df['party'] != 'democrat') & (elections_df['party'] != 'republican') & (elections_df['party'] != 'democratic-farmer-labor')]
elections_non_mp[elections_non_mp['candidate_evs_round'] > 0]
```

```python
calc_evs = elections_df[['year','state','candidate_evs_round']].groupby(['year','state']).sum().reset_index(drop=False)
calc_evs
```

```python
elections_df[(elections_df['year'] == 1992) & (elections_df['state'] == 'Georgia')]
```

```python
# get total number of actual evs per state per election year

ev_by_state = elections_df[['year','state','state_electoral_votes']].groupby(['year','state']).first().reset_index(drop=False)
ev_by_state
```

```python
ev_comp = pd.merge(calc_evs, ev_by_state, how='left', on=['year','state'])
ev_comp
```

```python
#if Match == False --> calculate match

def adjust_evs(row):
    if row['state_electoral_votes'] != row['candidate_evs_round']:
        if row['state_electoral_votes'] > row['candidate_evs_round']: 
            year = row['year']
            state = row['state']
            temp_df =  elections_df[(elections_df['year'] == year) & (elections_df['state'] == state)].reset_index(drop = True)
            temp_dict = {}
            total_evs = temp_df.loc[0,'state_electoral_votes']
            for index, row in temp_df.iterrows():
                candidate = row['candidate']
                candidate_evs = row['candidate_evs_round']
                calculated_evs = row['candidate_evs']
                distance = calculated_evs - candidate_evs
                temp_dict[candidate] = distance
            winner = max(temp_dict, key=temp_dict.get)
            row['candidate_evs_round'] += 1
            elections_df.loc[elections_df[(elections_df['year'] == year) & (elections_df['state'] == state) & (elections_df['candidate'] == winner)].index, 'candidate_evs_round'] += 1
            return adjust_evs(row)
        else:
            year = row['year']
            state = row['state']
            temp_df =  elections_df[(elections_df['year'] == year) & (elections_df['state'] == state)]
            temp_dict = {}
            total_evs = temp_df.loc[0,'state_electoral_votes']
            for index, row in temp_df.iterrows():
                candidate = row['candidate']
                candidate_evs = row['candidate_evs_round']
                calculated_evs = row['candidate_evs']
                distance = calculated_evs - candidate_evs
                temp_dict[candidate] = distance
            winner = min(temp_dict, key=temp_dict.get)
            row['candidate_evs_round'] -= 1
            elections_df.loc[elections_df[(elections_df['year'] == year) & (elections_df['state'] == state) & (elections_df['candidate'] == winner)].index, 'candidate_evs_round'] -= 1
            return adjust_evs(row)
    else:
        return

ev_comp['candidate_evs_round'] = ev_comp.apply(lambda x: adjust_evs(x), axis=1)

#for index, row in ev_comp.iterrows():
#    print(row['year'], row['state'])
```
